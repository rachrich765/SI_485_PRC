```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://ago.vermont.gov/archived-security-breaches/'
code = requests.get(url)
plain = code.text
s = BeautifulSoup(plain, "html.parser")
```

```python
year_urls= []
for x in s('li', {'class':"awDatesLI"}):
    for item in x('a'):
        year_urls.append(item.get('href'))
```

```python
url1= year_urls[1]
code = requests.get(url1)
plain = code.text
s = BeautifulSoup(plain, "html.parser")
```

```python
breaches = []
for x in s('div', {'id':"postWrapper"}):
    for item in x('a'):
        breaches.append(item.get('href'))

#breaches
```

```python
breach1= breaches[1]
code = requests.get(breach1)
plain = code.text
s = BeautifulSoup(plain, "html.parser")
```
